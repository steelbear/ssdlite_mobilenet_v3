```python
import torch
import torchvision
from torchvision.transforms import Resize, ToPILImage
from IPython.display import Image
```

```python
toPIL = ToPILImage()
resize = Resize((320, 320))
```

```python
image_5x5 = torchvision.io.read_image('circor/5x5/spec_16000_10/data/13918_AV_5.png')
toPIL(image_5x5)
```

```python
resized_image_5x5 = resize(image_5x5)
toPIL(resized_image_5x5)
```

```python
image_10x5 = torchvision.io.read_image('circor/spec_16000_10/data/13918_AV_5.png')
toPIL(image_10x5)
```

```python
resized_image_10x5 = resize(image_10x5)
toPIL(resized_image_10x5)
```
